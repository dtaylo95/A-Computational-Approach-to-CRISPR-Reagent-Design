#Using Google Colab

Google Colab is a platform that allows multiple users to view, edit, and run a shared Python notebook. A Python notebook is just what you're looking at now: it's a collection of text cells (like this one, called a "markdown" cell), and code cells, like the following:

````python
print('Hello')
````

You can "run" a code cell by clicking on it to select it, and then pressing `Ctrl + Enter`. Alternatively, you can just click the play button that appears on the left of the cell when your mouse hovers over it. Any output from running the code cell should show up below the code cell. Run the above code cell now, and take a look at the output of running the cell.

To edit a cell, you can double-click on it. Alternatively, when you have it selected, you can enter edit mode by pressing `Enter`.

To exit edit mode, you can press `Esc`.

Instead of clicking on cells to select them, you can also use the `up` and `down` arrow keys to move between cells. Just note that if you're in edit mode, using the arrow keys will move up and down within the cell.

If you want to create a new cell, you have two options:
1. You can press `Ctrl + M + B` to add a code cell *below* the cell you currently have selected, or
2. You can press `Ctrl + M + A` to add a code cell *above* the cell you currently have selected

Try creating a new cell below this cell.

If you want to add a new markdown cell, you'll need to create a new code cell first and then convert it to a markdown cell.

To convert a code cell to a markdown cell, you can press `Ctrl + M + M` when you have the code cell of interest selected.

To convert a markdown cell back to a code cell, you can press `Ctrl + M + Y` when you have the markdown cell of interest selected.

Convert the code cell you created above to a markdown cell.

If you want to delete a cell, you can press `Ctrl + M + D` when you have the cell of interest selected.

# Markdown Cells

Markdown cells are a great way of organizing your code into sections, and describing what your code does, both for others, as well as your future self.

For organization purposes, you can add headers using the hash `#` symbol. Enter edit mode for this markdown cell to see what that looks like.

There are multiple degrees of headers determined by the number of `#` you use. Take a look at the following.

# Header 1
## Header 2
### Header 3
#### Header 4
##### Header 5
###### Header 6

Using headers and subheaders will actually create a "Table of contents" that you can navigate. To view and navigate through the table of contents, click the following symbol at the top left of the screen.

![picture](https://drive.google.com/uc?export=view&id=1NLQPsGDwCukgl6tLns56Y8L3cKBl4TZI)

There are a ton of formatting options available in markdown text including: **bold** and *italics*. (Again enter edit mode for this cell to see how to do so.

You can also format text in a markdown cell as `code`. This won't run like a normal code cell, but it will look like one.

In addition to doing in-line code-formatted text, you can also create a block of text that is formatted like code, like this:

```
This is a code block.
It has multiple lines.
```

Note, the symbol used to do this is the block quote symbol paired with the tilda `~`  at the top left of your keyboard.

The last thing we're going to go over is lists: which you can do using the hyphen `-` symbol:
- Like this
- And this

Or you can make a numbered list using this `1.` format:
1. Like this
2. Or this

There are a lot of other ways you can format your markdown cells, and if you're interested in those, we encourage you to look them up yourself!
